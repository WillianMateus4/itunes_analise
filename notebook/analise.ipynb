```python
import pandas as pd
```

```python
df = pd.read_csv(r'..\dados\itunes_music_dataset.csv', sep=',')
```

```python
# Qual é o número total de músicas no dataset?
print(f"O dataset contém {len(df)} músicas.")

# Quais são os tipos de dados de cada coluna?
df.dtypes
```

```python
# Verifique as colunas country e currency. Elas possuem mais de um valor ou o dataset é exclusivo de um mercado específico (ex: USA / USD)?
parametros = ['country', 'currency']
df[parametros].value_counts()
```

```python
# Descubra quais colunas possuem valores nulos.
df.isnull().sum()
```

```python
# Criando coluna do artista principal
df['main_artist'] = df['artist_name']

# Padronizando os feats
df['main_artist'] = df['main_artist'].str.replace(' feat. ', ' & ', case=False)
df['main_artist'] = df['main_artist'].str.replace(' ft. ', ' & ', case=False)
df['main_artist'] = df['main_artist'].str.replace(' feat ', ' & ', case=False)
df['main_artist'] = df['main_artist'].str.replace(',', ' & ')

# Separa pelo ' & '
df['main_artist'] = df['main_artist'].str.split(' & ').str[0]

# Limpa espaços em brancos
df['main_artist'] = df['main_artist'].str.strip()
```

```python
# Ação: Preencha os preços nulos com o valor mediano do seu respectivo gênero musical. Exclua as linhas onde a data de lançamento (release_date) seja nula.
df['track_price'] = df.groupby('genre')['track_price'].transform(lambda x: x.fillna(x.median()))
df['collection_price'] = df.groupby('genre')['collection_price'].transform(lambda x: x.fillna(x.median()))
df = df.dropna(subset=['release_date'])
```

```python
df.info()
```

```python
# A coluna track_time_millis está em milissegundos, o que é péssimo para a leitura de negócios. Crie uma nova coluna chamada track_time_minutes convertendo esse valor para minutos.
df['track_time_minutes'] = round(df['track_time_millis'] / 60000, 2)
```

```python
df.head()
```

```python
# Converta a coluna release_date para o formato de Data (Datetime) e extraia uma nova coluna chamada release_year.
df['release_date'] = pd.to_datetime(df['release_date'])
df['release_year'] = df['release_date'].dt.year
```

```python
df.head()
```

```python
# Quais são os 5 gêneros musicais mais comuns neste dataset?
top_5_genre = df['genre'].value_counts().head(5).index
```

```python
# Qual é a duração média das músicas (track_time_minutes) para cada um desses 5 gêneros? (Ex: Músicas de Eletrônica são mais longas que as de Pop?)
df[df['genre'].isin(top_5_genre)].groupby('genre')['track_time_minutes'].mean().round(2)
```

```python
# Qual é o preço médio de uma música (track_price) e de um álbum (collection_price) geral?
parametros = ['track_price', 'collection_price']
df[parametros].mean().round(2)
```

```python
# Existe variação de preço dependendo do gênero? Construa uma tabela mostrando o track_price máximo, mínimo e médio por gênero.
df.groupby('genre').agg(
    máximo = ('track_price', 'max'),
    mínimo = ('track_price', 'min'),
    médio = ('track_price', 'mean')
    )
```

```python
# Quais são os 10 artistas com o maior número de faixas no dataset?
top_10_artist = df['artist_name'].value_counts().head(10).index
```

```python
# Desses 10 artistas, qual possui a maior média de track_price?
df[df['artist_name'].isin(top_10_artist)].groupby('artist_name')['track_price'].mean().round(2)
```

```python
# A duração das músicas está diminuindo com o passar dos anos? Agrupe os dados por release_year e calcule a média de track_time_minutes.
# (Dica de BI: filtre apenas os anos que possuem mais de 20 músicas lançadas para evitar distorções estatísticas de anos muito antigos).
contagem_anos = df['release_year'].value_counts()

filtros_anos = contagem_anos[contagem_anos > 20].index

df[df['release_year'].isin(filtros_anos)].groupby('release_year')['track_time_minutes'].mean().round(2)
```

```python
# A gravadora quer saber se vale a pena vender o álbum fechado ou apenas singles.
df_album = df.groupby('collection_name').agg(
    qtd_faixas = ('track_id', 'count'),
    preco_album = ('collection_price', 'mean'),
    soma_musica = ('track_price', 'sum'),
    track_price = ('track_price', 'mean')
)

# Filtra álbuns a partir de 5 músicas
df_album = df_album[df_album['qtd_faixas'] >= 5]

# Cria uma coluna para saber o desconto entre o álbum e a soma das músicas individuais do álbum
df_album['desconto_oferecido'] = df_album['soma_musica'] - df_album['preco_album']

# Exibi os 5 álbuns com o melhor desconto
df_album.sort_values(by='desconto_oferecido', ascending=False).head().round(2)
```

```python
# Descubra o número de faixas por álbum (collection_name) presentes no dataset.
df.groupby('collection_name')['track_id'].count().sort_values(ascending=False)
```

```python
# Para os álbuns que têm pelo menos 5 faixas neste dataset, calcule a métrica de "Custo por Faixa no Álbum" (collection_price / quantidade de faixas).
# Compare isso com o track_price médio do mesmo álbum. O iTunes dá desconto para quem compra o álbum completo?
df_album['Custo por Faixa no Álbum'] = df_album['preco_album'] / df_album['qtd_faixas']
df_album['Desconto Álbum por Faixa'] = df_album['Custo por Faixa no Álbum'] - df_album['track_price']

df_album.sort_values(by='Desconto Álbum por Faixa', ascending=True).head().round(2)
```
